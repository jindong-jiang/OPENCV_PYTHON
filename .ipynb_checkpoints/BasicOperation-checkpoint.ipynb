## Accessing Image Properties

```python
# numpy is optimized library for fast array calculations. Accessing each element using img[100,100] will be slow.
#Better use img.item ou img.itemset
img.item(10,10,2)
img.itemset((10,10,2),200)
img.item(10,10,2)
```

# Splitteing and Merging Image Channels

```python
# opencv gives a method but it is quite time costing
```

```python
b,g,r=cv2.split(img)
```

```python
plt.figure()
plt.imshow(b)
plt.figure()
plt.imshow(g)
plt.figure()
plt.imshow(r)
```

```python
b=img[:,:,0]
```

```python
img[:,:,2]=255 #much faster
```

```python
plt.imshow(img)
```

```python
img=cv2.merge((b,g,r))## Merge Image Channels|
```

## Making Borders for Images(Padding)

```python
BLUE=[255,0,0]
```

```python
for i,methode in enumerate([cv2.BORDER_REPLICATE,cv2.BORDER_REFLECT,cv2.BORDER_REFLECT_101,
         cv2.BORDER_WRAP,cv2.BORDER_CONSTANT]):
    
    pic_Padding=cv2.copyMakeBorder(img,50,50,50,50,methode)
    #plt.subplot(230+i+1)
    plt.figure(i)
    plt.imshow(pic_Padding)
```
