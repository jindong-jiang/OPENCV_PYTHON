## Smoothing Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## 2D Convolution

```python
#LPF: low pass filter for removing noises
#HPF: high pass filter for finding edges
```

```python
img=cv2.imread("pic3.jpg")
kernel=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)
```

```python
plt.imshow(img)
plt.figure()
plt.imshow(dst)
```

## Image Blurring(Image Smoothing)

```python
# Removing noises 
```

```python
# Averaging
blur=cv2.blur(img,(5,5))
plt.imshow(blur)
```

```python
blur=cv2.boxFilter(img,-1,(5,5),normalize=False)
plt.imshow(blur)
```

```python
blur=cv2.GaussianBlur(img,(5,5),1)
plt.imshow(blur)
```

```python
## Gaussian Blurring
cv2.getGaussianKernel(5,0)
```

```python
## Median Blurring
median=cv2.medianBlur(img,5)
plt.imshow(median)
```

```python
#Bilateral Filtering
blur=cv2.bilateralFilter(img,9,75,75)
plt.imshow(blur)
```
