## Geometric Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Scaling

```python
img=cv2.imread("pic3.jpg")
```

```python
plt.imshow(img)
```

```python
help(cv2.resize)
```

```python
res=cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
```

```python
plt.imshow(res)
```

```python
height,width=img.shape[:2]
res=cv2.resize(img,(2*width,2*height),interpolation=cv2.INTER_CUBIC)
plt.imshow(res)
#cv2.waitKey(5)
```

## Translation

```python
M=np.float32([[1,0,100],[0,1,50]])
rows,cols=img.shape[:2]
```

```python
dst=cv2.warpAffine(img,M,(2*cols,2*rows))
```

```python
plt.imshow(dst)
```

## Rotation

```python
M=cv2.getRotationMatrix2D((cols/2,rows/2),90,0.5)
```

```python
dst=cv2.warpAffine(img,M,(cols,rows))
```

```python
plt.imshow(dst)
```

## Affine Transformation

```python
pts1=np.float32([[50,50],[200,50],[50,200]])
pts2=np.float32([[10,100],[200,50],[100,250]])
```

```python
M=cv2.getAffineTransform(pts1,pts2)
```

```python
dst=cv2.warpAffine(img,M,(int(cols),int(rows)))
```

```python
plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(dst)
```

## Perspective Transformation

```python

pts2=np.float32([[10,100],[200,50],[100,250],[290,200]])
pts1=np.float32([[50,50],[200,50],[50,200],[200,200]])
```

```python
M=cv2.getPerspectiveTransform(pts2,pts1)
```

```python
dst2=cv2.warpPerspective(dst,M,(500,500))
```

```python
plt.subplot(311),plt.imshow(dst)
plt.subplot(312),plt.imshow(dst2)
plt.subplot(313),plt.imshow(img)
```
