```python
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("pic3.jpg",0)
```

```python
plt.imshow(img,cmap='gray',interpolation='bicubic')
plt.xticks([]), plt.yticks([])
```

```python
plt.imshow(cv2.line(img,(0,0),(511,511),(255,255,0),5),cmap="gray")
plt.imshow(cv2.rectangle(img,(110,110),(220,220),(0,255,0),2))
```

```python
font=cv2.FONT_HERSHEY_PLAIN
plt.imshow(cv2.putText(img,'picture',(0,300),font,5,(0,255,0),10,cv2.LINE_AA))
```
