## Morphological Transformations

```python
# Some simple operations based on the image shape
# Erosion Dialtion
# Opening Closing Gradient
```

## Erosion

```python
# Useful for removing the noise 
# Useful for detach two connected objects
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("j.png",0)
```

```python
kernel=np.ones((5,5),np.uint8)
```

```python
erosion=cv2.erode(img,kernel,iterations=1)
```

```python
plt.subplot(211),plt.imshow(img,cmap="gray")
plt.subplot(212),plt.imshow(erosion,cmap="gray")
```

## Dilation

```python
## Erosion is followed by dialtion
## After arosion, noises are gone
```

```python
dilatiion=cv2.dilate(img,kernel,iterations=1)
```

```python
plt.imshow(dilatiion,cmap="gray")
```

## Openning

```python
## Erosion+Dilation
```

```python
noise=np.array((np.random.rand(*img.shape)>0.5)*255,np.uint8)
```

```python
imgWithNoise=cv2.add(img,noise)
```

```python
plt.imshow(imgWithNoise,"gray")
```

```python
opening=cv2.morphologyEx(imgWithNoise,cv2.MORPH_OPEN,kernel)
```

```python
plt.imshow(opening,"gray")
```

## Closing

```python
# Dilation + Erosion
```

```python
#Useful in closing small holes inside the foreground boject
```

```python
imgWithNoise=img-noise
```

```python
plt.imshow(imgWithNoise,"gray")
```

```python
closing=cv2.morphologyEx(imgWithNoise,cv2.MORPH_CLOSE,kernel)
```

```python
plt.imshow(closing,"gray")
```

## Morphological Gradient

```python
# The difference between dilation and erosion of an image
```

```python
gradient=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
plt.imshow(gradient,"gray")
```

## Top Hat

```python
# difference between input image and Opening of the image
```

```python
tophat=cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernel)
```

```python
plt.imshow(tophat,"gray")
```

## Black Hat

```python
# Difference between the closing of the input and the input image
```

```python
blackhat=cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kernel)
```

```python
plt.imshow(blackhat,"gray")
```
