```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("pic2.jpg")
```

```python
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(img)
plt.figure()
plt.imshow(thresh1,cmap='gray')
```

## Adaptive Thresholding

```python
img=cv2.imread("pic2.jpg",0)
plt.imshow(img,cmap="gray")
```

```python
th2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                         cv2.THRESH_BINARY,11,2)
```

```python
plt.imshow(th2,"gray")
```

```python
th3=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
plt.imshow(th3,"gray")
```

## Otsu's Binarization

```python
Oret1,Oth1=cv2.threshold(th2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(Oth1,"gray")
```

```python
blur=cv2.GaussianBlur(th2,(5,5),0)
Oret2,Oth2=cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(Oth2,"gray")
```
